# Gold Layer Orchestration Notebook

Instead of running multiple notebooks manually, an orchestration notebook , runs transformations in the correct order , handling dependencies between datasets , and provides a single execution entry point. This notebook programmatically Validate Silver tables exist, Running dimension notebooks **first** , Running fact notebook **last** , and Ensure referential integrity
before Providing a single entry point for BI-ready data

### Validate Silver Layer Outputs
Validate the existence of each silver table before proceeding forward in order to Prevents Gold execution on incomplete Silver layer.

```python
tables = [
    "bike_lakehouse.silver.crm_customers" ,
    "bike_lakehouse.silver.crm_products",
    "bike_lakehouse.silver.crm_sales",
    "bike_lakehouse.silver.erp_customer_location",
    "bike_lakehouse.silver.erp_customers",
    "bike_lakehouse.silver.erp_product_category"
]
```

```python
for table in tables :

    if not spark.catalog.tableExists(table) :

        raise Exception(f"Missing required Bronze table: {table}") 
```

### Run Dimension Notebooks
Run dimension notebooks first to avoid any kind of dependencies 

```python
notebooks  = ["gold.dim_customers" , "gold.dim_products" ]
```

```python
for nb in notebooks :
    print(f"Executing notebook {nb} ...")
    dbutils.notebook.run(nb,timeout_seconds=0)
```

### Run Fact Notebook
Fact runs last to ensure dimensions exist, and successfully executed 

```python
print("Executing notebook gold.fact_sales ...")
dbutils.notebook.run("gold.fact_sales",timeout_seconds=0)
```
